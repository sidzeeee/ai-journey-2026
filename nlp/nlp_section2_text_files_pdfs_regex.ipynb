```python
from PyPDF2 import PdfReader

pdf_text = []

with open("US_Declaration.pdf", "rb") as f:
    pdf_reader = PdfReader(f)

    for page in pdf_reader.pages:
        pdf_text.append(page.extract_text())
```

```python
len(pdf_text)
```

```python
for page in pdf_text:
    print(page)
    print('\n')
```

## Regular Expressions

```python
text = "The phone number of the agent is 408-555-1243. Call soon!"
```

```python
"phone" in text
```

```python
import re
```

```python
pattern = "phone"
```

```python
re.search(pattern,text)
```

```python
my_match = re.search(pattern,text)
```

```python
my_match.span()
```

```python
text = "my phone is a new phone"
```

```python
match = re.search(pattern,text)
```

```python
match.span()
```

```python
all_matches = re.findall(pattern,text)
```

```python
len(all_matches)
```

```python
for match in re.finditer(pattern,text):
    print(match.span())
```

```python
text = "The phone number of the agent is 408-555-1243. Call soon!"
```

```python
text
```

```python
pattern = r"(\d{3})-(\d{3})-(\d{4})"
```

```python
phone_number = re.search(pattern,text)
```

```python
phone_number.span()
```

```python
phone_number.group(3)
```

```python
re.search(r"man | woman","This woman was here.")
```

```python
re.findall(r"..at","The cat in the hat sat splat.")
```

```python
re.findall(r"^\d","1,This number has ended 2")
```

```python
phrase = "There are 5 numbers and 55 in 30."
```
